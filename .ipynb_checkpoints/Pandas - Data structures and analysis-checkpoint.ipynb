```python
# Pandas - One-dimensional data (Values)

import pandas as pd
series = pd.Series([0,1,2,3,5,8,13], name='Results') ;series
```

```python
# Series creation: LIST

import numpy as np
an_array=np.arange(1,12, 2)
series=pd.Series(an_array, index = ['1st','2nd','3rd','4th','5th','6th']) ;series


print(an_array)
print('index is -> ', series.index)
print('values are -> ',series.values)
```

```python
# Series creation: DICTIONARY

a_dict = {'1st':1, '2nd':2, '3rd':3}
print('dictionary -> ', a_dict)

series=pd.Series(a_dict)
print(series)
```

```python
# Series creation: A SCALAR VALUE

series1=pd.Series(99)
print('series1 -> ',series1)

series2=pd.Series(99, index = ['1st', '2nd', '3rd', '4th', '5th', '6th'])
print('series2 -> ',series2)
```

```python
# Selecting elements from a Series
series=pd.Series(an_array, index = ['1st', '2nd', '3rd', '4th', '5th', '6th'])

print(series)
print(series[1])
print(series['3rd'])
```

```python
# Slicing elements from a Series

print(series[0:2])
```

```python
# Use a list of index numbers

series=series[[0,1]]
print(series)
```

```python
# Using conditions/operators

an_array=np.arange(1,12, 2)
series=pd.Series(an_array, index = ['1st','2nd','3rd','4th','5th','6th'])

series1=series[series>10] #bigger than @10
print('bigger than @10 == ', series1)

series2=series[np.logical_and(series>6,series<9)] #b/n than @6 and &10
print('#b/n than @6 and &10 == ', series2)
```

```python
# Using maths

series1=series+series
print(series1)

series2=series/2
print(series2)
```

```python
# DataFrame creation

an_array=np.array([(33,21), (24,57), (41,89)])
df = pd.DataFrame(an_array,columns=['age','marks']) ;df
```

```python
# DataFrame creation: Using a two-dimensional ndarray

df = pd.DataFrame(an_array, index=['mark', 'john', 'merry'], columns=['age','marks'])

print(df)
print(df.values)
print(df.index)
```

```python
url = "https://raw.githubusercontent.com/tvelichkovt/Scikit-Learn/main/WA_Fn-UseC_-Sales-Win-Loss.csv"
sales_data = pd.read_csv(url)
sales_data.head()
```

```python
# LOC -> to return rows indexed 0 and 1

sales_data.loc[[0,1]]
```

```python
# LOC -> to return rows indexed 0 and 1 with specific columns

sales_data.loc[[0,1], ['Region', 'Opportunity Result']]
```

```python
# ILOC -> integer based positions to access data instead of index and column labels

sales_data.iloc[1:4,0:3]
```

```python
# https://pandas.pydata.org/docs/user_guide/indexing.html
```

```python
# Summarizing Data

sales_data.head(n=2)
sales_data.tail(n=2)
```

```python
sales_data.describe()
```

```python
# return the number of rows that contain each of the unique values of variable

sales_data['Region'].value_counts()
```

```python
# return the number of unique values in this column

sales_data['Region'].nunique()
```

```python
# conditional selection

sales_data.groupby(['Region'])[['Opportunity Amount USD']].sum()
```

```python
# conditional selection

sales_data.groupby(['Region','Opportunity Result'])[['Opportunity Result']].count()
```

```python
sales_data1 = sales_data.copy()
sales_data1.describe()
```

```python
sales_data1.sort_values('Region', inplace = True)

filter1 = sales_data1['Region']=='Pacific'
filter2 = sales_data1['Opportunity Amount USD']>190000
filter3 = sales_data1['Opportunity Result']=='Won'

sales_data1.where(filter1 & filter2 & filter3, inplace = True)
sales_data1.to_csv('out.csv', index = None, header=True)

sales_data1.head()
```

```python
sales_data1.groupby(['Supplies Group'])[['Opportunity Amount USD']].sum()
```
